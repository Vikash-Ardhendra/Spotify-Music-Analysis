```python
#IN THIS DATASET, THERE ARE MANY TYPES OF GENRES. SO I AM GONNA CLASSIFY THESE GENRES IN SOME MAIN MUSIC GENRE
```

```python
df['genre'].unique()
```

```python
df['genre']=df['genre'].replace(['ambient','chill','new-age','comedy','club','party','show-tunes','singer-songwriter','songwriter','dance','romance','sad','sleep','emo'],
              'Other / Mood')
```

```python
df['genre']=df['genre'].replace(['rock','garage','alt-rock','hard-rock','punk','hardcore','punk-rock','metal','metalcore','psych-rock','grindcore','goth','rock-n-roll','progressive-house']
                               ,"Rock")
```

```python
df['genre']=df['genre'].replace(['electronic','electro','hardstyle','edm','techno','trance','deep-house','detroit-techno','dub','dubstep','minimal-techno','chicago-house','house','disco','breakbeat','drum-and-bass'],
                               'Electronic')
```

```python
df['genre'].unique()
```

```python
df['genre']=df['genre'].replace(['pop','pop-film','power-pop','k-pop','indie-pop'],'Pop')
```

```python
df['genre'].unique()
```

```python
df['genre']=df['genre'].replace(['afrobeat','cantopop','french','indian','swedish','spanish','german','ska','dancehall'],'World Music')
df['genre'].unique()
```

```python
df['genre']=df['genre'].replace(['folk','forro','salsa','samba','tango','sertanejo','acoustic'],'Folk')
df['genre'].unique()
```

```python
df['genre']=df['genre'].replace(['jazz'],"Jazz")
df['genre']=df['genre'].replace(['classical','opera'],'Classical')
df['genre']=df['genre'].replace(['blues','soul','gospel','groove'],"Blues & Soul")
df['genre']=df['genre'].replace(['hip-hop','trip-hop'],"Hip-Hop")
df['genre']=df['genre'].replace(['guitar','piano'],'Instrumental')
df['genre']=df['genre'].replace(['black-metal','death-metal','heavy-metal'],'Metal')
df['genre']=df['genre'].replace(['funk'],"Funk")
df['genre']=df['genre'].replace(['industrial'],'Industrial')
df['genre'].unique()
```
